We will start with discrete features 

Ploting the median value of SalePrice with respect to each value in the discrete features and also count of each value for better understanding the data.

```python
for feature in dis_features:
    plt.figure(figsize=(10,3))
    plt.subplot(1,2,1)
    train_df.groupby(feature)['SalePrice'].median().sort_values(ascending=False).plot.bar()
    plt.xlabel(feature)
    plt.ylabel('price')
    plt.subplot(1,2,2)
    train_df[feature].value_counts().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.show()
```

Now, we will see relation with 'SalePrice'

```python
for feature in con_features:
    sns.scatterplot(x=feature,y='SalePrice',data=train_df)
    plt.show()
```

From above,

        There are many features having more zero values, which is impacting the distribution of feature
        we are replacing these zero values with median imputation and capturing the missing value with extra variable.

```python
for feature in con_features:
    if len(train_df[train_df[feature]==0])>30:
        train_df[feature+'_Nan'] = np.where(train_df[feature]==0,1,0)
        random_sample = train_df[train_df[feature]!=0][feature].sample(len(train_df[train_df[feature]==0]), random_state=0,replace=True)
        # pandas needs to have the same index in order to merge datasets
        random_sample.index = train_df[train_df[feature]==0].index
        train_df.loc[train_df[feature]==0,feature] = random_sample
        
    
```

```python
for feature in con_features:
    if len(test_df[test_df[feature]==0])>30:
        test_df[feature+'_Nan'] = np.where(test_df[feature]==0,1,0)
        random_sample = test_df[test_df[feature]!=0][feature].sample(len(test_df[test_df[feature]==0]), random_state=0,replace=True)
        # pandas needs to have the same index in order to merge datasets
        random_sample.index = test_df[test_df[feature]==0].index
        test_df.loc[test_df[feature]==0,feature] = random_sample
```

```python
for feature in con_features:
    sns.scatterplot(x=feature,y='SalePrice',data=train_df)
    plt.show()
```

```python
for feature in cat_features:
    plt.figure(figsize=(10,3))
    plt.subplot(1,2,1)
    train_df.groupby(feature)['SalePrice'].median().sort_values(ascending=False).plot.bar()
    plt.xlabel(feature)
    plt.ylabel('price')
    plt.subplot(1,2,2)
    train_df[feature].value_counts().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.show()
```

As we can see we have fixed many categories with many labels into 4 to 5 labels only.

Now, we will encode the categorical features as well as discrete features using mean encoding

```python
def encode_categories(train_df,test_df,feature):
    ordered_labels = train_df.groupby(feature)['SalePrice'].mean().to_dict()
    train_df[feature] = train_df[feature].map(ordered_labels)
    test_df[feature] = test_df[feature].map(ordered_labels)
```

```python
for feature in cat_features:
    encode_categories(train_df,test_df,feature)
```

```python
for feature in dis_features:
    if feature not in ['MSSubClass', 'FullBath', 'TotRmsAbvGrd', 'Fireplaces','GarageCars']:
        encode_categories(train_df,test_df,feature)
```
