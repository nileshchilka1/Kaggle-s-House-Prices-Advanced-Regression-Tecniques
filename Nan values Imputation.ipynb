```python
train_df = pd.read_csv('train.csv')
```

First we check nan values

```python
train_df.isnull().sum()[train_df.isnull().sum()>0]
```

In Alley,PoolQC,Fence,MiscFeature contains more than 90% nan values.

we will drop these.

```python
test_df.isnull().sum()[test_df.isnull().sum()>0]
```

Here, also we can see there are many nan values.

```python
def drop_columns(train_df,test_df,columns):
    train_df.drop(columns,axis=1,inplace=True)
    test_df.drop(columns,axis=1,inplace=True)
```

we are also droping Id as it is not independent feature.

```python
drop_columns(train_df,test_df,['Id','Alley','PoolQC','Fence','MiscFeature'])
```

Now, we will start imputation in all the features one by one.

I have decided a threshold value in such way that if SalePrice is less than 250000 then I will use median imputation on it else I will impute 150 as LotFrontage.

```python
train_df['LotFrontage'] = np.where((train_df['LotFrontage'].isnull()) & (train_df['SalePrice']<250000),train_df['LotFrontage'].median(),train_df['LotFrontage'])
```

```python
train_df['LotFrontage'] = np.where((train_df['LotFrontage'].isnull()) & (train_df['SalePrice']>=250000),150,train_df['LotFrontage'])
```

```python
test_df['LotFrontage'] = np.where(test_df['LotFrontage'].isnull() ,train_df['LotFrontage'].median(),test_df['LotFrontage'])
```

For 'MasVnrType', I am using most frequent imputation

```python
train_df['MasVnrType'].value_counts()
```

```python
train_df['MasVnrType'] = np.where((train_df['MasVnrType'].isnull()),train_df['MasVnrType'].value_counts().index[0],train_df['MasVnrType'])
```

```python
test_df['MasVnrType'] = np.where(test_df['MasVnrType'].isnull(),train_df['MasVnrType'].value_counts().index[0],test_df['MasVnrType'])
```

As we can see if 'MasVnrType' is None then 'MasVnrArea' is 0.

so, we impute in same way.

```python
train_df[train_df['MasVnrType']=="None"]
```

```python
train_df['MasVnrArea'] = np.where((train_df['MasVnrArea'].isnull()) & (train_df['MasVnrType']=='None'),0,train_df['MasVnrArea'])
```

```python
test_df['MasVnrArea'] = np.where((test_df['MasVnrArea'].isnull()) & (test_df['MasVnrType']=='None'),0,test_df['MasVnrArea'])
```

For 'BsmtQual' I am using most frequent imputation

```python
train_df['BsmtQual'] = np.where((train_df['BsmtQual'].isnull()),train_df['BsmtQual'].value_counts().index[0],train_df['BsmtQual'])
```

```python
test_df['BsmtQual'] = np.where((test_df['BsmtQual'].isnull()),train_df['BsmtQual'].value_counts().index[0],test_df['BsmtQual'])
```

For 'BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','Electrical' I am using the same most frequent imputation.

```python
train_df['BsmtCond'] = np.where((train_df['BsmtCond'].isnull()),train_df['BsmtCond'].value_counts().index[0],train_df['BsmtCond'])
```

```python
test_df['BsmtCond'] = np.where((test_df['BsmtCond'].isnull()),train_df['BsmtCond'].value_counts().index[0],test_df['BsmtCond'])
```

```python
train_df['BsmtExposure'] = np.where((train_df['BsmtExposure'].isnull()),train_df['BsmtExposure'].value_counts().index[0],train_df['BsmtExposure'])
```

```python
test_df['BsmtExposure'] = np.where((test_df['BsmtExposure'].isnull()),train_df['BsmtExposure'].value_counts().index[0],test_df['BsmtExposure'])
```

```python
train_df['BsmtFinType1'] = np.where((train_df['BsmtFinType1'].isnull()),train_df['BsmtFinType1'].value_counts().index[0],train_df['BsmtFinType1'])
```

```python
test_df['BsmtFinType1'] = np.where((test_df['BsmtFinType1'].isnull()),train_df['BsmtFinType1'].value_counts().index[0],test_df['BsmtFinType1'])
```

```python
train_df['BsmtFinType2'] = np.where((train_df['BsmtFinType2'].isnull()),train_df['BsmtFinType2'].value_counts().index[0],train_df['BsmtFinType2'])
```

```python
test_df['BsmtFinType2'] = np.where((test_df['BsmtFinType2'].isnull()),train_df['BsmtFinType2'].value_counts().index[0],test_df['BsmtFinType2'])
```
